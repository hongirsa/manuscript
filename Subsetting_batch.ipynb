**Extract imageries based on tiles imported**

```
rootdir = '/content/drive/My Drive/FAPAR300'
parameter='FAPAR300'
for subdir, dirs, files in os.walk(rootdir):
  for dir in dirs:
    print ("Processing directory......:",dir)
    os.chdir(os.path.join(subdir, dir))
    tiles_gdf.MGRS_TILE
    for index, row in tiles_gdf.iterrows():      
      #print ("Processing tile...:",row.MGRS_TILE) 
      # print (row[['geometry']].values.flatten())
      try: 
        for filename in os.listdir(os.path.join(subdir, dir)):
          if filename.endswith("1.tiff"):
            with rasterio.open(filename) as src:
              out_image, out_transform = mask(src,row[['geometry']].values.flatten() , crop=True)      #clip file based on tile     
              out_meta = src.meta.copy()            
              out_meta.update({"driver": "GTiff",
                "height": out_image.shape[1],
                "width": out_image.shape[2],
                "transform": out_transform})
              file=os.path.splitext(filename)[0]          
              with rasterio.open(file + "_clipped.tif", "w", **out_meta) as dest:
                dest.write(out_image)
        #prepare the description for each file in merged file
        src_files_to_merge = []
        description=[]
        for filename1 in os.listdir(os.path.join(subdir, dir)):
          if filename1.endswith("clipped.tif"):
            #print (filename1)                 
            description.append(filename1[:filename1.index("_V1")])
            src_files_to_merge.append(filename1)              
        # Read metadata of first file
        with rasterio.open(src_files_to_merge[0]) as src0:
          meta = src0.meta
        # Update meta to reflect the number of layers
        meta.update(count = len(src_files_to_merge))
        # Read each layer and write it to stack
        with rasterio.open(os.path.join(subdir, dir)+"/"+parameter+"_"+dir+"_"+row.MGRS_TILE+".tiff", 'w', **meta) as dst:
          for id, layer in enumerate(src_files_to_merge, start=1):
            with rasterio.open(layer) as src1:
              dst.write_band(id, src1.read(1))   
              dst.set_band_description(id, description[id-1])
            #break
      except:
        print("Oops!  That was no valid imagery. ",row.MGRS_TILE)
      #remove temporay clipped file
      for filename in os.listdir(os.path.join(subdir, dir)):
        if filename.endswith("clipped.tif"):
          os.remove(filename)
```
